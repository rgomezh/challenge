> Creamos dos funciones para simplificar el c√≥digo

```python
# Function to make easier everything
def getNotNull(chunk):
    for col in chunk.columns:
        chunk=chunk[chunk[col].notnull()]     
    return chunk;
    
def getCleanUpperStr(chunk):
    for col in chunk.columns:
        if (chunk[col].dtypes == object):            
            chunk[col]=chunk[col].str.strip() 
            chunk[col]=chunk[col].str.upper() 
    return chunk
```

> #### Create booking_set, where we will have bookings with pax > 0, deduplicated by key with data "DepartureArribal_Date"

```python
# Create a list of new key with DepartureArribal_Date
bookings_set = pd.DataFrame()
for j, chunk_b in enumerate(
                    pd.read_csv(
                        bookings_bz,sep = '^', 
                        usecols = [6, 9, 12, 34], 
                        names = ['cre_date','dep_port','arr_port', 'pax' ], 
                        chunksize = 1000000,
                        #dtype={'pax': 'str','dep_port': 'str','arr_port': 'str'}
                        low_memory=False
                        )
                    ):
    chunk_b['cre_date'] = pd.to_datetime(chunk_b['cre_date'],errors='coerce')
    chunk_b['pax'] = pd.to_numeric(chunk_b['pax'], errors='coerce')
    
    # Filtering nulls
    chunk_b = getNotNull(chunk_b)
    
    # Bookings with pax < 0 are cancelations not bookings    
    chunk_b = chunk_b[chunk_b['pax'].astype(int)>0]      
    del chunk_b['pax']
    
    # Formating cols
    chunk_b = getCleanUpperStr(chunk_b)   
      
    # Create key for merging
    chunk_b['key'] = chunk_b['dep_port']+chunk_b['arr_port']+'_'+chunk_b['cre_date'].dt.strftime('%Y-%m-%d')

    # Delete columns not necessary yet
    del chunk_b['dep_port']
    del chunk_b['arr_port']
    del chunk_b['cre_date']  

    # Add new column for mark searches with bookings when we merge
    chunk_b['booking'] = '1'
    
    # Acumulation of new records
    bookings_set = bookings_set.append(chunk_b)
    print(j, len(chunk_b),len(bookings_set))
    
    # Delete duplicated bookings with the same key
    bookings_set = bookings_set.drop_duplicates()
    print(j, len(chunk_b),len(bookings_set))

    # Control for testing
    if j == 1:
        break
        

bookings_set
```

> #### Open Searches file to merge with bookins_set. First we need review data and create "key" to match columns

```python
# Merge Searches with grouped Bookings

for i, chunk_s in enumerate(
                    pd.read_csv(
                        searches_bz,sep = '^', 
                        low_memory = False,
                        chunksize = 1000000
                        )
                    ):
    searches_with_bookings = pd.DataFrame() 
    
    # Formating cols
    #chunk_s = getCleanUpperStr(chunk_s)
    chunk_s['Origin']=chunk_s['Origin'].str.strip().str.upper() 
    chunk_s['Destination']=chunk_s['Destination'].str.strip().str.upper()     
    
    # Create key to match with bookings
    chunk_s['key'] = chunk_s['Origin']+chunk_s['Destination']+'_'+chunk_s['Date']
    
    # Merge as left inner join. All searches + intersection with booking set.
    searches_with_bookings=chunk_s.merge(bookings_set, how='left')
    print(i, len(chunk_s))
    
    # All records with match with bookings the column booking field si 1. Others NaN. We change na by 0
    searches_with_bookings['booking'].fillna('0',inplace=True)             
    
    # Save partial blocks of searches to file. Thanks to mode we can append new lines to it. We compress because for size of file.
    # And we remove index field
    searches_with_bookings.to_csv(searches_bookings_bz, mode='a', index=False, sep = '^', compression='bz2')
    
    # Control for testing
    #if i == 2:
    #    break   

searches_with_bookings
        
    
```
